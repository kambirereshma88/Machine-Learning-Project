```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Scores.csv")     #linear_Regression always works for continious data  like price ,wt,age,distance,
df
```

```python
df.info()
```

```python
#Separation of x and y
x=df[["Hours"]]
y=df['Scores']
```

```python
x
```

```python
y
```

```python
#splitting into train and test
#      library.module                        class
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtrain
```

```python
ytrain
```

```python
#Build model    very important steps 
#step1:import model
from sklearn.linear_model import LinearRegression
#step2:create instance of model
lr=LinearRegression()
#step3:train a model
lr.fit(xtrain,ytrain)
#step4:Predict values
ypred=lr.predict(xtest)
```

```python
#r2score it one of the metrics to check acuracy
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
#to know best fit  line slope value 
lr.coef_
```

```python
#to known best fit line intercept
lr.intercept_
```
