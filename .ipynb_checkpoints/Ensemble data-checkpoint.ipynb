```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("bank_Dataset for DT.csv ")
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#splliting 

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

# Bagging

```python
#bagging

from sklearn.ensemble import BaggingClassifier
```

```python
bg=BaggingClassifier(LogisticRegression())
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
bg=BaggingClassifier(DecisionTreeClassifier())
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
from sklearn.ensemble import RandomForestClassifier    #by default it use DT
```

```python
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
ypred=rf.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
model=[]          #model is my list created 
accuracy=[]
model.append(("Logistic Regression",LogisticRegression()))
model.append(("DecisionTree",DecisionTreeClassifier()))
```

```python
model
```

```python
vc=VotingClassifier(estimators=model)
vc.fit(xtrain,ytrain)
ypred=vc.predict(xtest)
print(classification_report(ytest,ypred))
```

# Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada=AdaBoostClassifier()
ada.fit(xtrain,ytrain)
ypred=ada.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gd=GradientBoostingClassifier()
gd.fit(xtrain,ytrain)
ypred=gd.predict(xtest)
print(classification_report(ytest,ypred))
```
