```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
df=pd.read_excel('cust_data.xlsx')
```

```python
# Drop non-numeric columns and handle missing values
df_numeric = df.drop(['Cust_ID', 'Gender'], axis=1).fillna(0)
```

```python
# Standardize the data for better clustering
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)
```

```python
# Determine the optimal number of clusters using the elbow method
inertia = []
cluster_range = range(1, 11)
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
# Reduce the range of clusters and/or use a subset of data for faster processing
sample_df = df_scaled[:1000]  # Take a sample if the data is large
```

```python
# Re-run the elbow method with a reduced sample size
inertia_sample = []
for k in range(1, 6):  # Reduced range for faster processing
    kmeans_sample = KMeans(n_clusters=k, random_state=42)
    kmeans_sample.fit(sample_df)
    inertia_sample.append(kmeans_sample.inertia_)
```

```python
# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), inertia_sample, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters (Sample)')
plt.show()
```
