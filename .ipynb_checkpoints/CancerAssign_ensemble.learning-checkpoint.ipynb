```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv("cancer.csv ")
```

```python
# Step 4: Handle missing values (if any)
# Check for missing values
missing_values = df.isnull().sum()
```

```python
missing_values
```

```python
# Step 5: Exploratory Data Analysis (EDA)
# Exclude non-numeric columns (like 'diagnosis') for correlation
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns


# Check correlations between numeric variables
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 6: Data Preprocessing
# Handle categorical variables (if any) and encode labels
# Assuming the 'diagnosis' column is the target variable
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})  # 'M' -> Malignant, 'B' -> Benign
```

```python
# Separate features and target variable
x = df.drop(columns=['diagnosis'])  # Feature matrix (input data)
y = df['diagnosis']  # Target variable (output label)
```

```python
x
```

```python
y
```

```python
# Step 7: Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 8: Feature Scaling
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Step 9: Train Ensembling Models
# Initialize classifiers
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
gb_clf = GradientBoostingClassifier(n_estimators=100, random_state=42)
ab_clf = AdaBoostClassifier(n_estimators=100, random_state=42)
```

```python
# Train models
rf_clf.fit(x_train_scaled, y_train)
gb_clf.fit(x_train_scaled, y_train)
ab_clf.fit(x_train_scaled, y_train)
```

```python
# Step 10: Model Evaluation
# Predictions
rf_preds = rf_clf.predict(x_test_scaled)
gb_preds = gb_clf.predict(x_test_scaled)
ab_preds = ab_clf.predict(x_test_scaled)
```

```python
# Evaluation metrics
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, rf_preds))
print("\nGradient Boosting Classification Report:")
print(classification_report(y_test, gb_preds))
print("\nAdaBoost Classification Report:")
print(classification_report(y_test, ab_preds))
```

```python
# ROC AUC Score
rf_roc_auc = roc_auc_score(y_test, rf_clf.predict_proba(x_test_scaled)[:, 1])
gb_roc_auc = roc_auc_score(y_test, gb_clf.predict_proba(x_test_scaled)[:, 1])
ab_roc_auc = roc_auc_score(y_test, ab_clf.predict_proba(x_test_scaled)[:, 1])
```

```python
print(f"\nRandom Forest ROC-AUC: {rf_roc_auc:.2f}")
print(f"Gradient Boosting ROC-AUC: {gb_roc_auc:.2f}")
print(f"AdaBoost ROC-AUC: {ab_roc_auc:.2f}")
```

```python
# Confusion Matrix for Random Forest
plt.figure(figsize=(6, 6))
sns.heatmap(confusion_matrix(y_test, rf_preds), annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Step 11: Feature Importance (For Random Forest)
importances = rf_clf.feature_importances_
indices = np.argsort(importances)
```

```python
plt.figure(figsize=(10, 6))
plt.title("Random Forest - Feature Importances")
plt.barh(range(x.shape[1]), importances[indices], align="center")
plt.yticks(range(x.shape[1]), x.columns[indices])
plt.xlabel("Relative Importance")
plt.show()
```
