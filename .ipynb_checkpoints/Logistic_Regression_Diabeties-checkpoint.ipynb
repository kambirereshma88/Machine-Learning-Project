```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("diabeties.csv")
df
```

```python
df.info()
```

```python
df.describe()    #to check skewness use describe  
```

```python
#replace value of zero with nan then fill with mean value

for i in ['glocose','bp','skin','insulin','bmi']:
    df[i].replace(0,np.nan,inplace=True)
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
df.describe()
```

```python
df['target'].value_counts()
#this is imbalance data becoz its giving 0=500 ie no data is more than 1=268 
```

```python
#separation of x and y
x=df.iloc[:,:-1]
y=df['target']
```

```python
x
```

```python
y
```

```python
#split Data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#build model
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
#accc score 
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```

```python
#to check accuracy
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

```python
#hypertuning with solver
#solver=lbfgs  by default hyperparameter
#build model

lr=LogisticRegression(solver='liblinear')
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
lr=LogisticRegression(solver='newton-cg')
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
lr=LogisticRegression(solver='sag')
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
lr=LogisticRegression(solver='saga')    #sag and saga work on scalled data and give a good accuracy
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
print(classification_report(ytest,ypred))
```

```python
# if we want to know which acc is good so the acc which is highest ie if increase by hypertuning then take that acc as a higehst acc
#by default if acc is highest then that acc is best
```
