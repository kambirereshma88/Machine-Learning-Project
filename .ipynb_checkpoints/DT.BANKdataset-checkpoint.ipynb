```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("bank_Dataset for DT.csv ")
```

```python
df
```

```python
df.info()
```

```python
#separation of x and y

x=df.iloc[:,:-1]
y=df["deposit"]
```

```python
x
```

```python
y
```

```python
df['deposit'].value_counts()
```

```python

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
lr=LogisticRegression()    #create two object lr and dt
#svc=SVC()
dt=DecisionTreeClassifier()
```

```python
#calling all function

def mymodel(m):
    m.fit(xtrain,ytrain)
    ypred=m.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    cr=classification_report(ytest,ypred)
    print(f"Accuracy:  {ac} \n \n Classification Report: \n {cr}")
```

```python
mymodel(lr)
```

```python
mymodel(dt)
```

```python
#to check model overfitiing ahe ka
dt.score(xtrain,ytrain)  #testing score should be more  so this is overfitting problem
```

```python
dt.score(xtest,ytest)
```

```python
#hypertuning in DT ....1.max_depth
acc_list=[]
depth_list=[]
for i in range(1,16):
    dt1=DecisionTreeClassifier(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    ac=accuracy_score(ytest,ypred)

    depth_list.append(i)
    acc_list.append(ac)

    print(f"For max_depth={i}  We get Accuracy= {ac}")
```

```python
dt2=DecisionTreeClassifier(max_depth=7)
mymodel(dt2)
```

```python
dt2.score(xtrain,ytrain)
```

```python
#hypertuning in DT ....2.min_sample_split
acc_list=[]
n_sample_split_list=[]
for i in range(1,50):
    dt3=DecisionTreeClassifier(min_samples_split=2)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    ac=accuracy_score(ytest,ypred)

    n_sample_split_list.append(i)
    acc_list.append(ac)

    print(f"For max_depth={i}  We get Accuracy= {ac}")
```

```python
dt4=DecisionTreeClassifier(max_depth=48)
mymodel(dt4)
```

```python
#hypertuning in DT ....2.min_sample_leaf
acc_list=[]
min_sample_leaf_list=[]
for i in range(1,50):
    dt5=DecisionTreeClassifier(min_samples_leaf=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    ac=accuracy_score(ytest,ypred)

    min_sample_leaf_list.append(i)
    acc_list.append(ac)

    print(f"For min_sample_leaf={i}  We get Accuracy= {ac}")
```

```python
dt6=DecisionTreeClassifier(min_samples_leaf=40)
mymodel(dt6)
```

```python
dt6.score(xtrain,ytrain)
```
