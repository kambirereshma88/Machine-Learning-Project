```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("titanic_new.csv")
```

```python
df
```

```python
df.info()
```

```python
#to remove column  cabin from data use drop

df.drop("Cabin",axis=1,inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
#to replaace nan value with mean of 'Age'col use fill na method

df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.info()
```

```python
#to remove null values from column

df.dropna(inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df.drop(['PassengerId','Name','Ticket','Fare'],axis=1,inplace=True)

```

```python
df
```

```python
from sklearn.compose import ColumnTransformer    #coltransformer means to skip steps of coding and use cltransformer 
from sklearn.preprocessing import OrdinalEncoder

```

```python
#pass through mean to take all data with it 
ct=ColumnTransformer(transformers=[('Encoder',OrdinalEncoder(),['Sex','Embarked',])],remainder='passthrough') 
```

```python
x=df.iloc[:,1:]
y=df['Survived']
```

```python
x=ct.fit_transform(x)
```

```python
x[0]
```

```python
x
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
lr=LogisticRegression()
svc=SVC()
```

```python
def mymodel(model):
  model.fit(xtrain,ytrain)
  ypred=model.predict(xtest)
  print(classification_report(ytest,ypred))
  return model
```

```python
mymodel(lr)
```

```python
mymodel(svc)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
p1=Pipeline(steps=[('Scaler',StandardScaler()),('svm',SVC()),])
```

```python
p1.fit(xtrain,ytrain)
ypred=p1.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
plt.figure(figsize=(7,7))
sns.pairplot(df,hue='Survived')
plt.show()
```

```python
#it is a non_linear data
svm=SVC(kernel='linear')
mymodel(svm)
```

```python
svm=SVC(kernel='poly')
mymodel(svm)
```

```python
svm=SVC(kernel='sigmoid')
mymodel(svm)
```

```python
#hypertuning of c and gamma
from sklearn.model_selection import GridSearchCV
param_grid={'C':[1,10],'gamma':[1,10],'kernel':['rbf']}
```

```python
gd=GridSearchCV(svc,param_grid,verbose=3)#verbose (value)  is used to display information
gd.fit(xtrain,ytrain)                    #gridsearchcv is class   and param_grid=parameter
ypred=gd.predict(xtest)
```
