```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("cars.csv")
df
```

```python
x=df.iloc[:,:-1]
y=df['price']
```

```python
x
```

```python
y
```

```python
x.replace("?",np.nan,inplace=True)
```

```python
from sklearn.impute import SimpleImputer    #simpleImputer= replace nan with mean value   
si=SimpleImputer()
x[['normalized-losses','horsepower']]=si.fit_transform(x[['normalized-losses','horsepower']])
```

```python
from sklearn.preprocessing import OrdinalEncoder    #ordinal at atime replace all chara data into numerical data 
oe=OrdinalEncoder()
cat_col=x.select_dtypes(object).columns
x[cat_col]=oe.fit_transform(x[cat_col])
```

```python
x
```

```python
#splitting
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(xtrain,ytrain)

train=lreg.score(xtrain,ytrain)    #training Accuracy
test=lreg.score(xtest,ytest)       #testing Accuracy

print(f"Training Score:{train}")
print(f"Testing Score:{test}")
```

```python
 #the above values are showing training acc more than testing acc so it is overfitting model
```

```python
#Regularization ---------->Ridge(L2) and Lasso(L1)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
#L2 Regularization
l2=Ridge(alpha=5)  #alpha is a parameter f ridge(it is an error rate) 
l2.fit(xtrain,ytrain)   #programmatically lambda is known as alpha

train=l2.score(xtrain,ytrain)    #training Accuracy
test=l2.score(xtest,ytest)       #testing Accuracy

print(f"Training Score:{train}")
print(f"Testing Score:{test}")
```

```python
#Hypertuning of alpha value (mean to adjust value of hypermeter)
for i in range(1,30):
    l2=Ridge(alpha=i)
    l2.fit(xtrain,ytrain)
    test=l2.score(xtest,ytest)
    print(f"Value of lambda {i} Test Score{test}")
```

```python
l2=Ridge(alpha=11)  #alpha is a parameter f ridge(it is an error rate) 
l2.fit(xtrain,ytrain)   #programmatically lambda is known as alpha

train=l2.score(xtrain,ytrain)    #training Accuracy
test=l2.score(xtest,ytest)       #testing Accuracy

print(f"Training Score:{train}")
print(f"Testing Score:{test}")
```

```python
#by regularization will solve overfitting problem
```

```python
#L1Regularization --------->Lasso(L1)i.e Least Absolute and Selection Operator
#Hypertuning of alpha value (mean to adjust value of hypermeter)
for i in range(201,250):
    l1=Lasso(alpha=i)
    l1.fit(xtrain,ytrain)
    test=l1.score(xtest,ytest)
    print(f"Value of lambda {i} Test Score{test}")
```

```python
l1=Lasso(alpha=211)  #alpha is a parameter f ridge(it is an error rate) 
l1.fit(xtrain,ytrain)   #programmatically lambda is known as alpha

train=l1.score(xtrain,ytrain)    #training Accuracy
test=l1.score(xtest,ytest)       #testing Accuracy

print(f"Training Score:{train}")
print(f"Testing Score:{test}")
```

```python
#the above we solve overfitting problem in Lasso
```
