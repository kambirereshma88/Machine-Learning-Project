# KMeans implementation

```python
#kmean implementation

from sklearn.cluster import KMeans
```

### Elbow method 
To claculate best k(no.of Clusters)

```python
wcss=[]    # to save inertia wcss is taken like pandas as pd
for k in range(1,31):
    km=KMeans(n_clusters=k,init="k-means++",n_init=10,max_iter=300,random_state=1)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
plt.figure(figure=(6,5))
plt.title("Elbow Method")
plt.plot(range(1,31),wcss,marker='o')
plt.xlabel("k  Vlaues")
plt.ylabel("Inertia")
plt.show()
```

```python
km1=KMeans(n_clusters=5,random_state=1)
labels=km1.fit_predict(x)
```

```python
#PRinting inertia value for k=5
km1.inertia_
```

```python
km1.cluster_centers_
```

```python
labels
```

```python
df['c_labels']=labels
centroid_df=pd.DataFrame(km1.cluster_centers_,columns=['X','Y'])
```

```python
sns.scatterplot(data=df,x=df['Annual Income (k$)'],y=df['Spending Score (1-100)'],hue=df['c_labels'])
plt.scatter(centroid_df['X'],centroid_df['Y'],s=40,color='Yellow')
plt.show()
```
