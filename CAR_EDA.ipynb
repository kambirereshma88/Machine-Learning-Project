```python
#scaling : 1.standardScaler    2. MinMaxScaler
# To solve unit variation problems i.e to bring all data into one unit
```

```python
#MinMaxscaler ------> xnew=(x-xmin)/(xmax-xmin)
#it gives values between 0 and 1
```

```python
a=np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
c=np.array([100,200,300,400,500])
```

```python
def minmax(arr):              
    xnew=(arr-arr.min())/(arr.max()-arr.min())   # formula  =  (1-1)/(5-1) ,  (2-1)/(5-1)
    return xnew
```

```python
minmax(a)
```

```python
minmax(b)
```

```python
minmax(c)
```

```python
#from sklearn.preprocessing import MinMaxScaler
#standardScaler gives values betwn +infinity to -infinity 
#standardScaler-------> xnew=xmean/std Dev
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
feature.iloc[:,:]=sc.fit_transform(feature.iloc[:,:])
```
